**Exercício 1:**
Em um problema de classificação binária, a saída do modelo é a probabilidade de pertencimento à classe positiva. Qual distribuição você usaria para modelar essa saída? Justifique.


**Resposta:** A distribuição mais adequada para modelar a saída em problemas de classificação binária é a distribuição Binomial ou, mais especificamente, a distribuição Bernoulli quando consideramos uma única observação.

**Justificativa:** Pois a saída de um problema de classificação binária representa a probabilidade de pertencimento à classe positiva (por exemplo, P(Y=1)). Essa probabilidade está restrita ao intervalo [0,1], o que é característico da distribuição Bernoulli. Além disso, modelos como a regressão logística utilizam a função sigmoidal para mapear as saídas em probabilidades válidas no intervalo [0,1], assumindo implicitamente uma distribuição Binomial subjacente.

**Exercício 2:**
 Um algoritmo de clustering assume que os dados vêm de uma mistura de distribuições gaussianas. Que modelo estatístico isso descreve?

**Resposta:**
O modelo estatístico que descreve essa situação é a Mistura de Distribuições Gaussianas (Gaussian Mixture Model - GMM) .

**Justificativa:**
O GMM é um modelo probabilístico que assume que os dados são gerados por uma combinação linear de várias distribuições gaussianas. Cada componente gaussiano representa um cluster potencial nos dados, e o modelo ajusta os parâmetros das distribuições (média, covariância) e as proporções dos clusters para maximizar a verossimilhança dos dados. Algoritmos de clustering como o K-Means podem ser vistos como casos especiais do GMM, onde as covariâncias são diagonais e iguais.

**Exercício 3:**
 Suponha que você esteja treinando um modelo de machine learning com dados sintéticos gerados a partir de uma distribuição uniforme. Por que isso pode ser relevante para evitar viés nos dados?

**Resposta:**
Gerar dados sintéticos a partir de uma distribuição uniforme pode ser relevante para evitar viés nos dados por várias razões:

1-**Redução de viéses aos Dados Reais:**
Dados reais frequentemente apresentam padrões, tendências ou distribuições específicas que podem refletir viés na coleta ou na natureza dos dados. Ao usar uma distribuição uniforme, garante-se que não há preferência por certos valores ou regiões do espaço de características, o que pode ajudar a evitar que o modelo aprenda esses viéses.

2-**Teste de Generalização do Modelo:**
Dados uniformes permitem avaliar como o modelo se comporta em situações onde todas as possibilidades são igualmente prováveis. Isso ajuda a garantir que o modelo não depende excessivamente de padrões específicos presentes nos dados reais.

**3-Simetria e Equidade:**
Uma distribuição uniforme garante que todos os valores tenham a mesma probabilidade de ocorrer, o que pode ser útil para testar se o modelo trata todas as classes ou características de forma equitativa.
Simplificação do Problema:
Dados uniformes podem simplificar o ambiente experimental, permitindo focar exclusivamente no comportamento do modelo sem interferências complexas introduzidas por distribuições reais.

**Exercício 4:**
 Você está analisando o tempo entre cliques em anúncios online. Qual distribuição você usaria para modelar esse tempo? Por quê?

**Resposta:**
A distribuição mais adequada para modelar o tempo entre eventos independentes, como cliques em anúncios online, é a distribuição Exponencial .

**Justificativa:**
A distribuição exponencial é frequentemente usada para modelar tempos entre eventos em processos de Poisson, onde os eventos ocorrem de forma independente e aleatória ao longo do tempo. Ela captura bem a ideia de que o tempo até o próximo clique é independente do tempo decorrido desde o último clique. Além disso, sua propriedade de memória curta (a probabilidade de um evento ocorrer no próximo instante não depende do passado) é consistente com a natureza dos cliques online.


**Exercício 5:**
 Um sistema de recomendação coleta dados sobre o tempo gasto pelos usuários em páginas. Esses dados são assimétricos à direita. Qual distribuição seria adequada para modelá-los?

**Resposta:**
Uma distribuição adequada para modelar dados assimétricos à direita é a distribuição Log-Normal .

**Justificativa:**
Dados assimétricos à direita são comuns em fenômenos onde os valores pequenos são mais frequentes, mas existem alguns valores muito grandes que "puxam" a cauda para a direita. A distribuição Log-Normal é ideal para esses casos porque ela resulta de transformar uma variável normal em escala logarítmica, produzindo assim uma distribuição positiva e altamente assimétrica. Outras opções possíveis incluem a distribuição Weibull ou a distribuição Gamma, mas a Log-Normal é especialmente comum em contextos de tempo ou duração.

```python
#Exercício 6
#Enunciado: Use a distribuição de Poisson para simular o número de visitas a um site ao longo de 30 dias. Calcule a média e a variância dos dados simulados.
# Simulando 30 dias de visitas

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm

plt.style.use('seaborn-v0_8')
np.random.seed(42)  # Para reproduzibilidade
lambda_visitas = 10
num_dias = 30
visitas_simuladas = np.random.poisson(lam=lambda_visitas, size=num_dias)

# Mostrando as visitas simuladas
print("Visitas simuladas por dia:")
print(visitas_simuladas)

# Calculando média e variância empiricamente
media_empirica = np.mean(visitas_simuladas)
variancia_empirica = np.var(visitas_simuladas)

print(f"Média empírica: {media_empirica:.2f}")
print(f"Variância empírica: {variancia_empirica:.2f}")
```

```python
#Exercício 9:
#Enunciado: Use a biblioteca scipy.stats para calcular a probabilidade de obter 5 caras em 10 lançamentos de uma moeda justa.

#Resposta:
#Essa é uma questão de distribuição Binomial, pois estamos interessados no número de sucessos (caras) em um número fixo de tentativas (lançamentos).

from scipy.stats import binom

n = 10
p = 0.5
k = 5

probabilidade = binom.pmf(k, n, p)
print(f"Probabilidade de obter 5 caras em 10 lançamentos: {probabilidade:.4f}")
```
